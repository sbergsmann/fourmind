```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# make an array of x values from 0 to 100
C_e = np.arange(10, 255, 1)
C_p = np.arange(10, 255, 1)
C_p = 50

D = 0.15 * C_e + 0.36 * C_p - 0.0004 * C_e * C_p + 9.2
```

```python
# plot D vs. C_e
plt.figure()
plt.plot(C_e, D)
plt.xlabel('Previous message length (characters)')
plt.ylabel('Delay (s)')
plt.title('Delay vs. Previous message length')
plt.grid()
plt.show()
```
